```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

import pandas as pd
```

```python
data0 = pd.read_excel('HRDataset.xlsx')
```

```python
data0.head().transpose()
```

```python
subdata = data0.drop(["Employee_Name","EmpID","Termd","Position", "State", "Zip", "DOB", "Sex","MaritalDesc","RaceDesc","DateofHire","DateofTermination","TermReason","CitizenDesc","HispanicLatino","EmploymentStatus","Department","ManagerName","RecruitmentSource","PerformanceScore","LastPerformanceReview_Date","FromDiversityJobFairID","EngagementSurvey","ManagerID","DaysLateLast30"], axis = 1) 	
```

```python
subdata.head()
```

```python
scaler1=MinMaxScaler()

scaler1.fit(subdata[["MarriedID","MaritalStatusID","GenderID","EmpStatusID","DeptID","PerfScoreID","Salary","PositionID","EmpSatisfaction","SpecialProjectsCount","Absences"]])

subdata[["MarriedID","MaritalStatusID","GenderID","EmpStatusID","DeptID","PerfScoreID","Salary","PositionID","EmpSatisfaction","SpecialProjectsCount","Absences"]]=scaler1.transform(data0[["MarriedID","MaritalStatusID","GenderID","EmpStatusID","DeptID","PerfScoreID","Salary","PositionID","EmpSatisfaction","SpecialProjectsCount","Absences"]])

subdata.head().transpose()
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(subdata)

subdata['labels'] = kmeans.labels_
subdata['EmpID'] = data0['EmpID']
subdata['Salary'] = data0['Salary']
subdata.to_excel('HRClusteringResults2.xlsx', index=None)
```
