```python
import pandas as pd
file = open('AssignmentCopy.csv', 'r')

data1 = []

while 1:
    line = file.readline()
    if not line:
        break
    else:
        items= line.strip().split(',')
        has_item = {}
        for item in items:
            has_item[item] = 1
        data1.append(has_item)
file.close()

data = pd.DataFrame(data1)

data.fillna(0, inplace= True)
data.head(20)
```
