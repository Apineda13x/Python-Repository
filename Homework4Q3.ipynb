```python
import pandas as pd
data = pd.read_excel('Transactions01.xlsx')
data.fillna(0, inplace = True)
```

```python
from mlxtend.frequent_patterns import fpgrowth
freq_items = fpgrowth(data, min_support = 0.04, use_colnames = True, max_len=3)
freq_items.tail(10)
```

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(freq_items, metric = "lift", min_threshold = 1.5)
rules.to_excel('Transactions01.xlsx')
rules.head(10)
```

```python
rules.sort_values(by=['lift']).tail(10)
```
